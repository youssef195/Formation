```python
import numpy as np
import mediapipe as mp
import cv2  as cv

def camera():
    cap=cv.VideoCapture(0)
    if not cap.isOpened():
        print("peut pas ouvrir la camera")
        exit()
    while True:
        #frame par frame de image de la camera
        ret,frame=cap.read()
        
        #si il y'a une frame true
        if not ret:
            print("pas d'imge")
            break
        gray=cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        
        #l'image
        cv.imshow('frame',gray)
        if cv.waitKey(1)==ord('q'):
            break
    
    #fini
    cap.release()
    cv.destroyWindow()
camera
```

```python
# 3) c) L'histogramme du total d'argent dépensé UNIQUEMENT pour les comptes ayant dépensé de l'argent.
data_users_has_spent = data_users_viz[data_users_viz.has_ordered == int(True)].copy()
fig = px.histogram(data_users_has_spent, x="account_id", y="money_spent", title="L'histogramme du total d'argent dépensé UNIQUEMENT pour les comptes ayant dépensé de l'argent.")
fig.show()
```

```python
# 4) a) Pareil pas le souvenir d'avoir vu ça en cours mdrrr
data_clients = data_users[data_users.has_ordered == int(True)].copy()
data_clients = data_clients[data_clients.last_connection_since >= 30].reset_index(drop=True)
data_clients.head()
```

```python
# 4) b) Changer la phrase (le texte en rouge entre les guillemets)
def texte_mail(account_id):
    client = data_clients.loc[data_clients['account_id'] == account_id]
    discount = '10' if client['money_spent'].values <= 50 else '20'
    return "Bonjour utilisateur " + str(account_id) + " ! Pour récompenser votre fidélité, nous vous offrons " + discount + " euros de réductions sur votre prochaine commande." 
```
